```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
import numpy as np

# Create figure and axis with a specified size
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_aspect('equal')  # Ensure circles appear circular

# Define parameters
r = 1.0  # Radius of each circle
d = 1.5  # Distance between centers (adjusted for smaller intersections)

# Define the centers of the three circles in an equilateral triangle formation
centers = [
    (0, 0),              # Bottom-left circle
    (d, 0),              # Bottom-right circle
    (d / 2, d * np.sqrt(3) / 2)  # Top circle
]

# Draw the circles
for center in centers:
    circle = Circle(
        center,
        r,
        facecolor='lightgray',  # Light gray fill
        edgecolor='black',     # Black outline
        alpha=0.5,             # Transparency for visible overlaps
        lw=2                   # Line width of outline
    )
    ax.add_patch(circle)

# Define label positions within the non-intersecting areas of each circle
label_positions = [
    (-0.4, 0),              # Inside left circle (Financial management)
    (d + 0.4, 0),           # Inside right circle (Investment management)
    (d / 2, d * np.sqrt(3) / 2)  # Inside top circle (Capital markets)
]

# Define the labels with newlines for multi-line text
labels = [
    "Financial\nmanagement",
    "Investment\nmanagement",
    "Capital markets\nand capital market theory"
]

# Add labels to the plot
for pos, label in zip(label_positions, labels):
    ax.text(
        pos[0], pos[1],
        label,
        ha='center',          # Center horizontally
        va='center',          # Center vertically
        fontsize=10,          # Font size
        fontweight='bold',    # Bold text
        color='black'         # Black text
    )

# Set plot limits to display the entire diagram
ax.set_xlim(-1.5, d + 1.5)
ax.set_ylim(-1.5, d * np.sqrt(3) / 2 + 1.5)

# Remove axes for a clean look
ax.axis('off')

# Display the plot
plt.show()
```
