```python
import matplotlib.pyplot as plt

# Data setup
years = [0, 1, 2, 3, 4, 5]
cash_flow = 5000
discount_rate = 0.07

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Draw red timeline
ax.hlines(y=0, xmin=-0.5, xmax=5.5, colors='red', linewidth=2)

# Add vertical ticks for each year
ax.vlines(x=years, ymin=-0.1, ymax=0.1, colors='black')

# Label the years below the timeline
for year in years:
    ax.text(year, -0.3, f'Year {year}', ha='center', va='top')

# Add PV formula labels above Years 1 to 5 as fractions
for i in range(1, 6):
    formula = f'$\\frac{{5,000}}{{(1+0.07)^{{{i}}}}}$'
    ax.text(i, 0.5, formula, ha='center', va='bottom')

# Add arrows and cash flow labels from each year to Year 0
for i in range(1, 6):
    ax.annotate(
        f'${cash_flow}',
        xy=(0, 0),  # Point to Year 0
        xytext=(i, 0),  # Start from each year
        arrowprops=dict(
            arrowstyle='->',
            connectionstyle='arc3,rad=0.2',
            color='black',
            linestyle='dotted'
        ),
        ha='center',
        va='center',
        xycoords='data',
        textcoords='data',
        bbox=dict(boxstyle='round,pad=0.2', fc='white', ec='none')
    )

# Add discount rate note above the timeline
#ax.text(2.5, 1.5, 'Discount rate = 7%', ha='center', va='center', fontsize=10)

# Customize the plot
ax.set_xlim(-0.5, 5.5)
#ax.set_ylim(-1.5, 2.0)
ax.set_ylim(-2.0, 2.0)
ax.set_yticks([])  # Hide y-axis ticks
ax.set_xticks([])  # Hide x-axis ticks
ax.set_ylabel('Present value (year 0)', fontsize=10)

# Remove spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# Display the plot
plt.tight_layout()
plt.show()
```
