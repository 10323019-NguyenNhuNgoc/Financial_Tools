```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrowPatch, ArrowStyle

def draw_timeline(ax):
    """
    Draws the timeline from 0 to 3, showing step-by-step
    deposits and compounding, ending at $3641.
    """
    # Hide default axes, set plot range
    ax.axis('off')
    ax.set_xlim(-0.5, 3.5)
    ax.set_ylim(-1.0, 0.2)
    
    # Draw timeline at y=0, with ticks at x=0,1,2,3
    ax.hlines(y=0, xmin=0, xmax=3, color='black')
    for x in [0, 1, 2, 3]:
        ax.vlines(x, ymin=-0.03, ymax=0.03, color='black')
        ax.text(x, 0.06, str(x), ha='center', va='bottom', fontsize=9)  # label above the line
    
    # Title
    ax.text(1.5, 0.15, "Timeline Figure", ha='center', va='bottom', fontsize=11)
    
    # Use an arrow style for consistent arrow heads
    arrow_style = ArrowStyle("->", head_length=6, head_width=3)
    
    #----------------------------------------------------------------
    # CHAIN OF PARTIAL SUMS (vertical/horizontal steps)
    #----------------------------------------------------------------
    # 1) $1000 at t=0
    ax.text(0, -0.1, "$1000", ha='center', va='top', fontsize=10)
    
    # Arrow from (0, -0.1) to (1, -0.1) => $1100 at (1, -0.1)
    arrow1 = FancyArrowPatch((0, -0.1), (1, -0.1),
                             connectionstyle="arc3,rad=0",
                             arrowstyle=arrow_style, color='gray')
    ax.add_patch(arrow1)
    ax.text(0.5, -0.07, "×1.10", ha='center', va='bottom', fontsize=9)
    ax.text(1, -0.1, "$1100", ha='center', va='top', fontsize=10)
    
    # Arrow from (1, -0.1) down to (1, -0.25) => $2100
    arrow2 = FancyArrowPatch((1, -0.1), (1, -0.22),
                             connectionstyle="arc3,rad=0",
                             arrowstyle=arrow_style, color='gray')
    ax.add_patch(arrow2)
    ax.text(1, -0.25, "$2100", ha='center', va='top', fontsize=10)
    
    # Arrow from (1, -0.25) to (2, -0.25) => $2310
    arrow3 = FancyArrowPatch((1, -0.25), (2, -0.25),
                             connectionstyle="arc3,rad=0",
                             arrowstyle=arrow_style, color='gray')
    ax.add_patch(arrow3)
    ax.text(1.5, -0.22, "×1.10", ha='center', va='bottom', fontsize=9)
    ax.text(2, -0.25, "$2310", ha='center', va='top', fontsize=10)
    
    # Arrow from (2, -0.25) down to (2, -0.4) => $3310
    arrow4 = FancyArrowPatch((2, -0.25), (2, -0.38),
                             connectionstyle="arc3,rad=0",
                             arrowstyle=arrow_style, color='gray')
    ax.add_patch(arrow4)
    ax.text(2, -0.4, "$3310", ha='center', va='top', fontsize=10)
    
    # Arrow from (2, -0.4) to (3, -0.4) => $3641
    arrow5 = FancyArrowPatch((2, -0.4), (3, -0.4),
                             connectionstyle="arc3,rad=0",
                             arrowstyle=arrow_style, color='gray')
    ax.add_patch(arrow5)
    ax.text(2.5, -0.37, "×1.10", ha='center', va='bottom', fontsize=9)
    ax.text(3, -0.4, "$3641", ha='center', va='top', fontsize=10)

# Create a single figure and axis
fig, ax = plt.subplots(figsize=(8, 4))

# Draw the timeline on the single axis
draw_timeline(ax)

plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrowPatch, ArrowStyle

def draw_annuity_timeline(ax):
    """
    Draws a timeline (0..3) with three separate $1000 deposits:
    - Deposit at t=0 -> grows to $1331 at t=3
    - Deposit at t=1 -> grows to $1210 at t=3
    - Deposit at t=2 -> grows to $1100 at t=3
    Final total at t=3 is $3641.
    """
    # Hide default axes; adjust the visible region
    ax.axis('off')
    ax.set_xlim(-0.5, 3.5)
    ax.set_ylim(-0.8, 0.3)

    # ------------------------------------------------------------------
    # 1) Draw the main timeline at y=0, with ticks at x=0,1,2,3
    # ------------------------------------------------------------------
    ax.hlines(y=0, xmin=0, xmax=3, color='black', lw=1)
    for x in [0, 1, 2, 3]:
        ax.vlines(x, ymin=-0.03, ymax=0.03, color='black', lw=1)
        ax.text(x, 0.06, str(x), ha='center', va='bottom', fontsize=9)  # label 0,1,2,3 above the line

    # Optional: A title
    ax.text(1.5, 0.2, "Multiple Deposits at 10% Interest", ha='center', va='bottom', fontsize=11)

    # Define an arrow style for consistent arrow heads
    arrow_style = ArrowStyle("->", head_length=6, head_width=3)

    # ------------------------------------------------------------------
    # 2) TOP FLOW: Deposit at t=0 -> $1331 at t=3
    # ------------------------------------------------------------------
    y_top = -0.1
    # a) Label $1000 at t=0
    ax.text(0, y_top, "$1000", ha='center', va='top', fontsize=10)

    # b) Arrow from t=0 to t=1 (×1.10), label at midpoint, $1100 at t=1
    arrow1 = FancyArrowPatch((0.05, y_top), (0.95, y_top),
                             connectionstyle="arc3,rad=0",
                             arrowstyle=arrow_style, color='gray')
    ax.add_patch(arrow1)
    ax.text(0.5, y_top + 0.03, "×1.10", ha='center', va='bottom', fontsize=9)
    ax.text(1, y_top, "$1100", ha='center', va='top', fontsize=10)

    # c) Arrow from t=1 to t=2, $1210 at t=2
    arrow2 = FancyArrowPatch((1.05, y_top), (1.95, y_top),
                             connectionstyle="arc3,rad=0",
                             arrowstyle=arrow_style, color='gray')
    ax.add_patch(arrow2)
    ax.text(1.5, y_top + 0.03, "×1.10", ha='center', va='bottom', fontsize=9)
    ax.text(2, y_top, "$1210", ha='center', va='top', fontsize=10)

    # d) Arrow from t=2 to t=3, $1331 at t=3
    arrow3 = FancyArrowPatch((2.05, y_top), (2.95, y_top),
                             connectionstyle="arc3,rad=0",
                             arrowstyle=arrow_style, color='gray')
    ax.add_patch(arrow3)
    ax.text(2.5, y_top + 0.03, "×1.10", ha='center', va='bottom', fontsize=9)
    ax.text(3, y_top, "$1331", ha='center', va='top', fontsize=10)

    # ------------------------------------------------------------------
    # 3) MIDDLE FLOW: Deposit at t=1 -> $1210 at t=3
    # ------------------------------------------------------------------
    y_mid = -0.3
    # a) Label $1000 at t=1
    ax.text(1, y_mid, "$1000", ha='center', va='top', fontsize=10)

    # b) Arrow from t=1 to t=2, $1100 at t=2
    arrow4 = FancyArrowPatch((1.05, y_mid), (1.95, y_mid),
                             connectionstyle="arc3,rad=0",
                             arrowstyle=arrow_style, color='gray')
    ax.add_patch(arrow4)
    ax.text(1.5, y_mid + 0.03, "×1.10", ha='center', va='bottom', fontsize=9)
    ax.text(2, y_mid, "$1100", ha='center', va='top', fontsize=10)

    # c) Arrow from t=2 to t=3, $1210 at t=3
    arrow5 = FancyArrowPatch((2.05, y_mid), (2.95, y_mid),
                             connectionstyle="arc3,rad=0",
                             arrowstyle=arrow_style, color='gray')
    ax.add_patch(arrow5)
    ax.text(2.5, y_mid + 0.03, "×1.10", ha='center', va='bottom', fontsize=9)
    ax.text(3, y_mid, "$1210", ha='center', va='top', fontsize=10)

    # ------------------------------------------------------------------
    # 4) BOTTOM FLOW: Deposit at t=2 -> $1100 at t=3
    # ------------------------------------------------------------------
    y_bot = -0.5
    # a) Label $1000 at t=2
    ax.text(2, y_bot, "$1000", ha='center', va='top', fontsize=10)

    # b) Arrow from t=2 to t=3, $1100 at t=3
    arrow6 = FancyArrowPatch((2.05, y_bot), (2.95, y_bot),
                             connectionstyle="arc3,rad=0",
                             arrowstyle=arrow_style, color='gray')
    ax.add_patch(arrow6)
    ax.text(2.5, y_bot + 0.03, "×1.10", ha='center', va='bottom', fontsize=9)
    ax.text(3, y_bot, "$1100", ha='center', va='top', fontsize=10)

    # ------------------------------------------------------------------
    # 5) Final sum of all three deposits at t=3 => $3641
    # ------------------------------------------------------------------
    # We'll place it just below the bottom flow
    ax.text(3, -0.65, "$3641", ha='center', va='top', fontsize=10)
    # Draw a short horizontal line to mimic underlining
    ax.hlines(y=-0.66, xmin=2.93, xmax=3.07, color='black', lw=1)


# ----------------------------------------------------------------------
# Create a figure and draw the timeline with multiple deposits
# ----------------------------------------------------------------------
fig, ax = plt.subplots(figsize=(9, 4))
draw_annuity_timeline(ax)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrowPatch, ArrowStyle

def draw_multiple_deposits(ax):
    """
    Draw a timeline (0..3) at the top, with three separate $1000 deposits:
      1) Deposit at t=0 -> grows to $1331 at t=3
      2) Deposit at t=1 -> grows to $1210 at t=3
      3) Deposit at t=2 -> grows to $1100 at t=3
    The final total at t=3 is $3641 (underlined).

    All arrows are shortened by 10% compared to the previous code.
    """
    # Hide default axes; set the visible region
    ax.axis('off')
    ax.set_xlim(-0.5, 3.5)
    ax.set_ylim(-0.8, 0.3)

    # ------------------------------------------------------------------
    # 1) Main timeline at y=0, with ticks at x=0,1,2,3
    # ------------------------------------------------------------------
    ax.hlines(y=0, xmin=0, xmax=3, color='black', lw=1)
    for x in [0, 1, 2, 3]:
        ax.vlines(x, ymin=-0.03, ymax=0.03, color='black', lw=1)
        ax.text(x, 0.06, str(x), ha='center', va='bottom', fontsize=9)

    # Optional title
    ax.text(1.5, 0.2, "Multiple Deposits at 10% Interest",
            ha='center', va='bottom', fontsize=11)

    # Define an arrow style
    arrow_style = ArrowStyle("->", head_length=6, head_width=3)

    # Helper function to shorten an arrow by 10%:
    #   If old arrow is (start, end) with length 0.9,
    #   we cut 0.045 from each side => new length 0.81.
    def shorten_arrow(x_start, x_end):
        # Original length
        length = x_end - x_start
        # Reduce length by 10%
        new_length = 0.9 * length
        # Split the "lost" 10% equally from start and end
        trim = (length - new_length) / 2
        return x_start + trim, x_end - trim

    # ------------------------------------------------------------------
    # 2) TOP FLOW: deposit at t=0 -> grows to $1331 by t=3
    # ------------------------------------------------------------------
    y_top = -0.1
    ax.text(0, y_top, "$1000", ha='center', va='top', fontsize=10)

    # Arrow: 0→1
    x1a, x1b = shorten_arrow(0.05, 0.95)
    arrow0_1 = FancyArrowPatch((x1a, y_top), (x1b, y_top),
                               connectionstyle="arc3,rad=0",
                               arrowstyle=arrow_style, color='gray')
    ax.add_patch(arrow0_1)
    ax.text((x1a + x1b) / 2, y_top + 0.03, "×1.10", ha='center', va='bottom', fontsize=9)
    ax.text(1, y_top, "$1100", ha='center', va='top', fontsize=10)

    # Arrow: 1→2
    x2a, x2b = shorten_arrow(1.05, 1.95)
    arrow1_2 = FancyArrowPatch((x2a, y_top), (x2b, y_top),
                               connectionstyle="arc3,rad=0",
                               arrowstyle=arrow_style, color='gray')
    ax.add_patch(arrow1_2)
    ax.text((x2a + x2b) / 2, y_top + 0.03, "×1.10", ha='center', va='bottom', fontsize=9)
    ax.text(2, y_top, "$1210", ha='center', va='top', fontsize=10)

    # Arrow: 2→3
    x3a, x3b = shorten_arrow(2.05, 2.95)
    arrow2_3 = FancyArrowPatch((x3a, y_top), (x3b, y_top),
                               connectionstyle="arc3,rad=0",
                               arrowstyle=arrow_style, color='gray')
    ax.add_patch(arrow2_3)
    ax.text((x3a + x3b) / 2, y_top + 0.03, "×1.10", ha='center', va='bottom', fontsize=9)
    ax.text(3, y_top, "$1331", ha='center', va='top', fontsize=10)

    # ------------------------------------------------------------------
    # 3) MIDDLE FLOW: deposit at t=1 -> grows to $1210 by t=3
    # ------------------------------------------------------------------
    y_mid = -0.3
    ax.text(1, y_mid, "$1000", ha='center', va='top', fontsize=10)

    # Arrow: 1→2
    x4a, x4b = shorten_arrow(1.05, 1.95)
    arrow1_2_mid = FancyArrowPatch((x4a, y_mid), (x4b, y_mid),
                                   connectionstyle="arc3,rad=0",
                                   arrowstyle=arrow_style, color='gray')
    ax.add_patch(arrow1_2_mid)
    ax.text((x4a + x4b) / 2, y_mid + 0.03, "×1.10", ha='center', va='bottom', fontsize=9)
    ax.text(2, y_mid, "$1100", ha='center', va='top', fontsize=10)

    # Arrow: 2→3
    x5a, x5b = shorten_arrow(2.05, 2.95)
    arrow2_3_mid = FancyArrowPatch((x5a, y_mid), (x5b, y_mid),
                                   connectionstyle="arc3,rad=0",
                                   arrowstyle=arrow_style, color='gray')
    ax.add_patch(arrow2_3_mid)
    ax.text((x5a + x5b) / 2, y_mid + 0.03, "×1.10", ha='center', va='bottom', fontsize=9)
    ax.text(3, y_mid, "$1210", ha='center', va='top', fontsize=10)

    # ------------------------------------------------------------------
    # 4) BOTTOM FLOW: deposit at t=2 -> grows to $1100 by t=3
    # ------------------------------------------------------------------
    y_bot = -0.5
    ax.text(2, y_bot, "$1000", ha='center', va='top', fontsize=10)

    x6a, x6b = shorten_arrow(2.05, 2.95)
    arrow2_3_bot = FancyArrowPatch((x6a, y_bot), (x6b, y_bot),
                                   connectionstyle="arc3,rad=0",
                                   arrowstyle=arrow_style, color='gray')
    ax.add_patch(arrow2_3_bot)
    ax.text((x6a + x6b) / 2, y_bot + 0.03, "×1.10", ha='center', va='bottom', fontsize=9)
    ax.text(3, y_bot, "$1100", ha='center', va='top', fontsize=10)

    # ------------------------------------------------------------------
    # 5) Final sum at t=3: $3641 (underlined)
    # ------------------------------------------------------------------
    final_sum_y = -0.7
    ax.text(3, final_sum_y, "$3641", ha='center', va='top', fontsize=10)
    # Short horizontal line to mimic underlining
    #ax.hlines(y=final_sum_y - 0.01, xmin=2.93, xmax=3.07, color='black', lw=1)


# ----------------------------------------------------------------------
# Create the figure and draw the single timeline with shorter arrows
# ----------------------------------------------------------------------
fig, ax = plt.subplots(figsize=(9, 4))
draw_multiple_deposits(ax)
plt.tight_layout()
plt.show()


```
